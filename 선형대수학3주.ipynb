```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([4,3])
#Figure 객체를 만든 후
fig=plt.figure()
##1x1 개의 서브플롯을 추가, 첫 번째 서브플롯을 사용
ax=fig.add_subplot(1,1,1)

#변수 a의 값을 좌표로 사용하여 점을 그린다.
#a[0]이 x좌표, a[1]이 y좌표, s는 점의 크기
ax.scatter(a[0],a[1],s=30)
#점 옆에 문자 a를 표시, sive는 문자의 크기
ax.text(a[0]+0.2, a[1]+0.2, 'a', size=15)
#x축, y축 눈금 범위를 -5~5로 지정
ax.set_xticks(range(-5,6))
ax.set_yticks(range(-5,6))
#그리드를 좌표축 아래에 표시
ax.grid()
ax.set_axisbelow(True)
#x축과 y축 눈금의 비율을 1:1로 함
#창의 크기를 변경해도 비율이 유지됨
ax.set_aspect('equal', adjustable='box')
#다음 코드를 사용하여 원점에 좌표축이 보이게함
#왼쪽 y축과 아래 x축을(0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

화살표 그리기

```python
a = np.array([4,3])
##Figure 객체를 만든 후
fig=plt.figure()
#1x1개의 서브 플롯을 추가, 첫 번째 서브플롯을 사용
ax=fig.add_subplot(1,1,1)

#a[0]과 a[1]이 각각 끝 위치의 x좌표, y좌표
ax.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1)
ax.text(a[0],a[1], 'a', size=15)
#x축 y축 눈금 범위를 -5~5로 지정
ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
#그리드를 좌표축 아래에 표시
ax.grid()
ax.set_axisbelow(True)
#x축과 y축 눈금의 비율을 1:1로 함
#창의 크기를 변경해도 비율이 유지됨
ax.set_aspect('equal', adjustable='box')
#다음 코드를 사용하여 원점에 좌표축이 보이게함
#왼쪽 y축과 아래 x축을(0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

다수의 서브플롯 사용하기

```python
#벡터 A와 B를 생성
A=np.array([1,-2])
B=np.array([4,3])

#1개의 행과 2개의 열을 가지는 서브플롯을 생성
f,ax=plt.subplots(1,2)
ax[0].title.set_text('A')
ax[1].title.set_text('B')

#각 서브플롯에 시작점이 (0,0)인 벡터를 그린다.
ax[0].quiver(0,0,A[0],A[1],angles='xy',scale_units='xy',scale=1)
ax[1].quiver(0,0,B[0],B[1],angles='xy',scale_units='xy',scale=1)

#서브플롯별로 그리드를 생성
start_x =-1
end_x = 5
start_y = -3
end_y = 4
for i in range(2):
  ax[i].axis([start_x,end_x,start_y,end_y])
  ax[i].set_xticks(range(start_x, end_x))
  ax[i].set_yticks(range(start_y,end_y))
  ax[i].grid(True)
  ax[i].set_aspect('equal', adjustable='box')

plt.show()

```

3차원 공간에 화살표 그리기

```python
#3차원벡터 A
A = np.array([4,5,6])
fig=plt.figure()

#서브플롯 dptj 3차원 좌표계를 사용
ax=fig.add_subplot(1,1,1,projection='3d')

#시작점이 (0,0,0)이고 끝점 좌표가 변수 a값인 화살표 그리기
ax.quiver(0,0,0,A[0],A[1],A[2],color='black',arrow_length_ratio=0.1)
ax.text(A[0],A[1],A[2], 'a',size=15)

#그리드 생성
ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20.,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto', adjustable='box')
plt.show()
```

벡터의 기본 연산(덧셈)

```python
a = np.array([4,1])
b = np.array([-2,3])
sum = a+b

fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#벡터a를 원점에서 시작하도록 그린다.
ax.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='blue')
ax.text(a[0],a[1],'a',size=15)

#벡터b를 벡터a의 끝점에서 시작하도록 그린다.
ax.quiver(a[0],a[1],b[0],b[1],angles='xy',scale_units='xy',scale=1,color='green')
ax.text(a[0]+b[0],a[1]+b[1],'b',size=15)

ax.quiver(0,0,sum[0],sum[1],angles='xy',scale_units='xy',scale=1,color='red')
ax.text(sum[0]*0.5+0.3,b[1],'a+b',size=15, color='blue')

ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

#왼쪽 y축과 아래 x축을(0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

뺄셈

```python
a = np.array([4,1])
b = np.array([-2,3])
sum = a-b

fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#벡터a를 원점에서 시작하도록 그린다.
ax.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='blue')
ax.text(a[0],a[1],'a',size=15)

#벡터b를 벡터a의 끝점에서 시작하도록 그린다.
ax.quiver(a[0],a[1],-b[0],-b[1],angles='xy',scale_units='xy',scale=1,color='green')
ax.text(a[0]-b[0],a[1]-b[1],'b',size=15)

ax.quiver(0,0,sum[0],sum[1],angles='xy',scale_units='xy',scale=1,color='red')
ax.text(sum[0]*0.5+0.3,b[1],'a+b',size=15, color='blue')

ax.set_xticks(range(-2,8))
ax.set_yticks(range(-4,5))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

#왼쪽 y축과 아래 x축을(0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

스칼라곱

```python
a = np.array([4,1])
c =3
ca = np.multiply(c,a)

fig=plt.figure()
ax = fig.add_subplot(1,1,1)
#벡터a를 원점에서 시작하도록 그린다.
ax.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='blue')
ax.text(a[0],a[1],'a',size=15)

#벡터 a에 3을 곱한 결과를 그린다.
ax.quiver(0,1,ca[0],ca[1],angles='xy',scale_units='xy',scale=1,color='green')
ax.text(ca[0],ca[1],'3a',size=15, color='blue')

ax.set_xticks(range(-2,15))
ax.set_yticks(range(-3,9))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

#왼쪽 y축과 아래 x축을(0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

내적

```python
a=np.array([1,2])
b=np.array([2,3])
print(np.dot(a,b))

print(np.linalg.norm(a))
print(np.sqrt(5))

#자기자신에 대한 내적의 제곱근
print(np.sqrt(np.dot(a,a)))
#벡터의 크기
print(np.linalg.norm(a))
```

외적

```python
from mpl_toolkits.mplot3d import Axes3D

a= np.array([5,3,0])
b = np.array([1,2,0])
#외적 계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#a벡터를 그린다.
ax.quiver(0,0,0,a[0],a[1],a[2],color='black', arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)

#b벡터를 그린다.
ax.quiver(0,0,0,b[0],b[1],b[2],color='red', arrow_length_ratio=0.1)
ax.text(b[0],b[1],b[2],'b',size=15)

#벡터a와 벡터b의 외적을 그린다.
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue', arrow_length_ratio=0.1)
ax.text(c[0],c[1],c[2],'c',size=15)

#그리드 생성
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20.,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto', adjustable='box')
plt.show()
```

```python
a= np.array([3,3,0])
b = np.array([6,6,0])
#외적 계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#a벡터를 그린다.
ax.quiver(0,0,0,a[0],a[1],a[2],color='black', arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)

#b벡터를 그린다.
ax.quiver(0,0,0,b[0],b[1],b[2],color='red', arrow_length_ratio=0.1)
ax.text(b[0],b[1],b[2],'b',size=15)

#벡터a와 벡터b의 외적을 그린다.
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue', arrow_length_ratio=0.1)
ax.text(c[0],c[1],c[2],'c',size=15)

#그리드 생성
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20.,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto', adjustable='box')
plt.show()
```

행렬의 정의

```python
A=np.array([[1,2,3,4],[5,6,7,8]])
#행 벡터 출력
print(A[0])
print(A[0,:])
print(A[1])
print(A[1,:])
#열 벡터 출력
print(A[:,0])
print(A[:,1])
print(A[:,2])
print(A[:,3])
```

```python
#행렬의 덧셈, 뺄셈
A=np.array([[1,2],[3,4]])
B=np.array([[4,1],[5,2]])

print(A+B)
print(np.add(A,B))
print(A-B)
print(np.subtract(A,B))
print()

#행렬의 스칼라 곱
A=np.arange(9).reshape(3,3)
print(A)
c=10
print(np.multiply(c,A))
print(c*A)
```
